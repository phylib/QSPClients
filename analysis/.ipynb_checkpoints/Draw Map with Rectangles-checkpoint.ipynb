```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import matplotlib.collections as coll
import matplotlib.text as text
from matplotlib.gridspec import GridSpec
import matplotlib.image as mpimg
import os

#plt.rcParams.update({'font.size': 22})

def setParams():
    mpl.rcParams['pdf.fonttype'] = 42
    mpl.rcParams['ps.fonttype'] = 42
    #plt.rcParams.update({'font.size': 14})
    mpl.rcParams['font.size'] = 12
```

```python
wid = 8
hei = 8
nrows = 8
ncols = 8

rectW = wid / float(ncols)
rectH = hei / float(nrows)

xx = np.arange(0, wid, rectW)
yy = np.arange(0, hei, rectH)

def drawRect(ax, label, (x, y), width, height, bgColor="white", edgecolor='black', 
             linestyle='-', linewidth=1.0, hatch=None, fontweight='regular'):
    sq = patches.Rectangle((x, y), width, height, facecolor=bgColor, edgecolor=edgecolor, 
                           linestyle=linestyle, linewidth=linewidth, hatch=hatch)
    ax.add_artist(sq)
    if label is not None:
        t = text.Text(x + (width / 2.), y + (height / 2.), label, ha='center', va='center', axes=ax, weight=fontweight)
        ax.add_artist(t)

def drawCircle(ax, label, (x, y), radius, bgColor="white", borderColor="red", linestyle='-'):
    
    # ax1.add_patch(patches.Circle((0.5, 0.5),
    #                         radius=0.5,
    #                         color='k', linewidth=10, fill=False))
    
    circle = patches.Circle((x, y), radius=radius, facecolor=bgColor, 
                           edgecolor=borderColor, linestyle=linestyle,
                           linewidth=2)
    ax.add_artist(circle)
    if label is not None:
        t = text.Text(x, y, label, ha='center', va='center', axes=ax)
        ax.add_artist(t)
    
    
def draw_server_chart(ax):
    ax.set_aspect('equal')
    s_num = 0
    for yi in yy:
        for xi in xx:
            if s_num == 0 or s_num == 27:
                drawRect(ax, "R" + str(s_num), (xi, yi), rectW, rectH, bgColor="lightgrey")
            else:
                drawRect(ax, "R" + str(s_num), (xi, yi), rectW, rectH)
            s_num += 1

    ax.set_xlim(-0.3, wid + 0.3)
    ax.set_ylim(-0.3, hei + 0.3)
    ax.axis("off")
    ax.set_title('a) The game world is divided\ninto 64 smaller regions.', y=-0.17)
    
    
def draw_lu_server(ax):
    ax.set_aspect('equal')
    s_num = 0
    drawRect(ax, None, (0, 0), 8, 8)
    #drawRect(ax, "P0", (0, 0), 1, 1, bgColor='lightgrey')
    
    
    drawRect(ax, u"", (1, 0), 1, 1, bgColor='lightblue')
    drawRect(ax, u"", (0, 1), 1, 1, bgColor='lightblue')
    drawRect(ax, u"", (1, 1), 1, 1, bgColor='lightblue')
    
    drawRect(ax, u"", (2, 0), 2, 2)
    drawRect(ax, u"", (0, 2), 2, 2)
    drawRect(ax, u"", (2, 2), 2, 2)
    
    drawRect(ax, u"", (4, 0), 4, 4)
    drawRect(ax, u"", (0, 4), 4, 4)
    drawRect(ax, u"", (4, 4), 4, 4)
    
    drawCircle(ax, "P0", (0.5, 0.5), radius=0.7, bgColor='lightgreen', linestyle='--')
    
    s_num += 1

    ax.set_xlim(-0.3, wid + 0.3)
    ax.set_ylim(-0.3, hei + 0.3)
    ax.axis("off")
    ax.set_title('b) Peer P0, responsible for region\nR0, divides the game world\ninto 10 regions.', y=-0.24)
    
def draw_center_server(ax):
    ax.set_aspect('equal')
    s_num = 0
    drawRect(ax, None, (0, 0), 8, 8)
    #drawRect(ax, "P27", (3, 3), 1, 1, bgColor='lightgrey')
    
    
    drawRect(ax, u"", (2, 2), 1, 1, bgColor='lightblue')
    drawRect(ax, u"", (2, 3), 1, 1, bgColor='lightblue')
    drawRect(ax, u"", (3, 2), 1, 1, bgColor='lightblue')
    drawRect(ax, u"", (3, 4), 1, 1, bgColor='lightblue')
    drawRect(ax, u"", (4, 3), 1, 1, bgColor='lightblue')
    drawRect(ax, u"", (4, 4), 1, 1, bgColor='lightblue')
    drawRect(ax, u"", (4, 2), 1, 1, bgColor='lightblue')
    drawRect(ax, u"", (2, 4), 1, 1, bgColor='lightblue')
    
    drawRect(ax, u"", (5, 2), 1, 1)
    drawRect(ax, u"", (5, 3), 1, 1)
    drawRect(ax, u"", (5, 4), 1, 1)
    drawRect(ax, u"", (5, 5), 1, 1)
    drawRect(ax, u"", (2, 5), 1, 1)
    drawRect(ax, u"", (3, 5), 1, 1)
    drawRect(ax, u"", (4, 5), 1, 1)
    
    drawRect(ax, u"", (0, 0), 2, 2)
    drawRect(ax, u"", (0, 2), 2, 2)
    drawRect(ax, u"", (0, 4), 2, 2)
    drawRect(ax, u"", (0, 6), 2, 2)
    drawRect(ax, u"", (2, 0), 2, 2)
    drawRect(ax, u"", (4, 0), 2, 2)
    drawRect(ax, u"", (6, 0), 2, 2)
    
    drawRect(ax, u"", (2, 6), 2, 2)
    drawRect(ax, u"", (4, 6), 2, 2)
    drawRect(ax, u"", (6, 6), 2, 2)
    drawRect(ax, u"", (6, 2), 2, 2)
    drawRect(ax, u"", (6, 4), 2, 2)
    
    drawCircle(ax, "P1", (3.5, 3.5), radius=0.7, bgColor='lightgreen', linestyle='--')
    
    s_num += 1

    ax.set_xlim(-0.3, wid + 0.3)
    ax.set_ylim(-0.3, hei + 0.3)
    ax.axis("off")
    ax.set_title('c) Peer P1, responsible for region\nR27, divides the game world\ninto 28 regions.', y=-0.24)
    

fig, axes = plt.subplots(1, 3, figsize=(15, 5), dpi=90)
setParams()
draw_server_chart(axes[0])
draw_lu_server(axes[1])
draw_center_server(axes[2])

#plt.tight_layout()
plt.savefig("peerEvaluationScenario.pdf")
os.system("pdfcrop --margins '5 5 5 5' peerEvaluationScenario.pdf")
```

```python
def plot_no_knowledge(ax):
    draw_all_rects(ax)

    drawRect(ax, "Local region", (0, 4), 4, 4)
    
    drawRect(ax, "Remote Region:\n\nNo information", (4, 4), 4, 4, edgecolor='lightgrey', hatch='/')
    drawRect(ax, "Remote Region:\n\nNo information", (4, 0), 4, 4, edgecolor='lightgrey', hatch='/')
    drawRect(ax, "Remote Region:\n\nNo information", (0, 0), 4, 4, edgecolor='lightgrey', hatch='/')
    drawRect(ax, None, (4, 4), 4, 4, bgColor='#00000000')
    drawRect(ax, None, (4, 0), 4, 4, bgColor='#00000000')
    drawRect(ax, None, (0, 0), 4, 4, bgColor='#00000000')
    
    ax.set_title("1) Initial State: No knowlege\nabout remote regions", y=-0.17)
    
def plot_step_1(ax):
    draw_all_rects(ax)

    drawRect(ax, "Local region", (0, 4), 4, 4)
    
    drawRect(ax, "Remote Region:\n\nHash " + u"", (4, 4), 4, 4, bgColor='#adebad')
    drawRect(ax, "Remote Region:\n\nHash " + u"", (0, 0), 4, 4, bgColor='#adebad')
    drawRect(ax, "Remote Region:\n\nHash " + u"", (4, 0), 4, 4, bgColor='#ebbcad')
    
    
    #drawRect(ax, None, (4, 0), 2, 2)
    #drawRect(ax, None, (4, 2), 2, 2)
    #drawRect(ax, None, (6, 0), 2, 2)
    #drawRect(ax, None, (6, 2), 2, 2)
    
    ax.set_title("2) Received hash values for first\nlevel", y=-0.17)
    
def plot_step_2(ax):
    draw_all_rects(ax)

    drawRect(ax, "Local region", (0, 4), 4, 4)
    
    drawRect(ax, "Remote Region:\n\nHash " + u"", (4, 4), 4, 4, bgColor='#adebad')
    drawRect(ax, "Remote Region:\n\nHash " + u"", (0, 0), 4, 4, bgColor='#adebad')
    #drawRect(ax, "Remote Region:\n\nHash changed", (4, 0), 4, 4, bgColor='#ebbcad')
    
    
    drawRect(ax, "Hash " + u"", (6, 0), 2, 2, bgColor='#adebad')
    drawRect(ax, "Hash " + u"", (6, 2), 2, 2, bgColor='#adebad')
    drawRect(ax, "Hash " + u"", (4, 0), 2, 2, bgColor='#adebad')
    
    
    drawRect(ax, u"", (4, 3), 1, 1, bgColor='#ebbcad')
    drawRect(ax, u"", (5, 2), 1, 1, bgColor='#ebbcad')
    drawRect(ax, u"", (4, 2), 1, 1, bgColor='#adebad')
    drawRect(ax, u"", (5, 3), 1, 1, bgColor='#adebad')
    
    ax.set_title("3) Result of recursive hash comparison\nup to the third tree level", y=-0.17)
    
fig, axes = plt.subplots(1, 3, figsize=(13, 5), dpi=90)
setParams()
plt.rcParams["hatch.linewidth"] = 8

plot_no_knowledge(axes[0])
plot_step_1(axes[1])
plot_step_2(axes[2])

plt.tight_layout()
plt.savefig("gaining-knowledge.pdf")
os.system("pdfcrop --margins '5 5 5 5' gaining-knowledge.pdf")
```
